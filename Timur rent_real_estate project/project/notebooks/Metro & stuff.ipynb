```python
s='м. Преображенская площадь (13 мин пешком)'
```

```python
def metro(s):
    s=s[3:]
    for i in range(0,len(s)):
        if s[i]=='(':
            s=s[0:i-1]
            break
    s.strip()
    return s
```
