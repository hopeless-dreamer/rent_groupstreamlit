```python
import pandas as pd 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
mpl.pyplot.bar(fd2.columns,
               fd2.isnull().sum(),
               width=0.4,
               color='orange',
               bottom=None,
               align='center')
plt.xticks(rotation=90, ha='center')
plt.xlabel('Столбцы')
plt.ylabel('Количество пустых значений')
plt.title('Пустые значения в столбцах');
```

По строкам пустые на подавляющее большинство квартир приходится от 3 до 7 пропусков в значениях. И значения отсутствуют в половине полей у подавляющего меньшенства арендодателей. 

```python
plt.style.use('ggplot')
plt.hist(fd2.isnull().sum(axis=1),
               width=1,
               color='brown')
plt.xticks(rotation=90, ha='center')
plt.xlabel('Количество пустых значений в строке')
plt.title('Распределение количества пустых значений по строкам при 25-и столбцах');
```
